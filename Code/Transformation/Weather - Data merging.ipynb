```python
import xarray as xr
import pandas as pd
import os
from pathlib import Path
```

# Loading data

## Load data from SSRD and temperature

```python

# Load the NetCDF file
base_path = Path(os.getcwd())

file_relative = r"..\..\Data\Raw\reanalysis-era5-single-levels-timeseries-sfc7rupio5o.nc"

file_path = (base_path / file_relative).resolve()

ds = xr.open_dataset(file_path)

# Inspect the dataset first
print("=== VARIABLES ===")
print(ds.data_vars)
print("\n=== COORDINATES ===")
print(ds.coords)
print("\n=== DIMENSIONS ===")
print(ds.dims)
```

```python
# Convert to pandas DataFrame
df_bol = ds.to_dataframe().reset_index()

# Basic inspection
print(df_bol.head())
print(df_bol.dtypes)
print(df_bol.shape)
```

```python
df_bol['SSRD_Wm2'] = df_bol['ssrd'] / 3600

df_bol['date'] = df_bol['valid_time'].dt.date
df_bol['hour'] = df_bol['valid_time'].dt.time

#rename valid_time to time
df_bol.rename(columns={'valid_time': 'time'}, inplace=True)
```

```python
df_bol.head(15)
```

## Load SSRD Clear sky

```python
# Load the NetCDF file
base_path = Path(os.getcwd())

file_relative = r"..\..\Data\Raw\cams_solar_irradiance_ts.nc"

file_path = (base_path / file_relative).resolve()

df_cs = xr.open_dataset(file_path)
```

```python
# Convert to pandas DataFrame
df_bol_cs = df_cs.to_dataframe().reset_index()

# Basic inspection
print(df_bol_cs.columns)
print(df_bol_cs.dtypes)
```

```python
# Selecting 'time', 'latitude', 'longitude', 'CLEAR_SKY_GHI', 'GHI' from df_bol_cs
df_bol_ghi = df_bol_cs[['time', 'CLEAR_SKY_GHI', 'GHI']]
df_bol_ghi.head(15)
```

## Joining dataframes

```python
#Drop latitude and longitude from df_s

# Join df_bol and df_selected on 'time', 'latitude', 'longitude'
df_bol_full = pd.merge(df_bol, df_bol_ghi, on=['time'], how='inner')

# select only the columns 'time', 'latitude', 'longitude', 'SSRD_Wm2', 'CLEAR_SKY_GHI', 'GHI'
df_bol_full = df_bol_full[['time', 'latitude', 'longitude', 't2m', 'SSRD_Wm2', 'CLEAR_SKY_GHI', 'GHI']]

df_bol_full.head(15)
```

### Saving the final dataframe

```python

base_path = Path(os.getcwd())

file_relative = r"..\..\Data\Raw\df_bologna_full.parquet"

output_path = (base_path / file_relative).resolve()

df_bol_full.to_parquet(output_path, index=False)
```
